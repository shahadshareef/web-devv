```python
def my_num(number):
    if number % 2 == 0:
        return("this number is even")
    else:
        return("this number is odd")
    
my_num(78)
```

```python
def isPalindrome(string):
    if string == string[::-1]:
        return True
    else:
        return False
    

isPalindrome("racecar")
```

```python
for i in range(1,6):
    for j in range(1,6):
        if(j>=i):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
